# Zeitverschiebung

Manchmal musst du vielleicht Deine kompletten Daten entlang des Zeitreihenindex verschieben (Time Shifting). Tatsächlich tun das viele der eingebauten Funktionen von Pandas bereits. Auch wenn wir so etwas in diesem Kurs nicht oft machen werden, ist es definitiv gut, davon zu wissen!

```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('time_data/walmart_stock.csv',index_col='Date')
df.index = pd.to_datetime(df.index)
```

```python
df.head()
```

```python
df.tail()
```

## .shift() vorwärts

```python
df.shift(1).head()
```

```python
# Du wirst den letzten Teil deiner Daten, der keinen Index mehr hat, verlieren!
df.shift(1).tail()
```

## shift() rückwärts

```python
df.shift(-1).head()
```

```python
df.shift(-1).tail()
```

## Shifting basierend auf einem Zeit-Stringcode
### Mithilfe von tshift()

```python
# Verschiebe alles einen Monat nach vorn
df.tshift(periods=1,freq='M').head()
```
